```python
import pandas as pd
akas = pd.read_csv("../../truncated_data/truncated_title.akas.tsv", sep='\t')
```

```python
with open('../../../SQL/population/done/aka_insertions.sql','w') as f:
    for index, row in akas.iterrows():
        title_id = row['titleId'] if row['titleId'] != r'\N' else "NULL"
        ordering = row['ordering'] if row['ordering'] != r'\N' else "NULL"
        title = row['title'] if row['title'] != r'\N' else "NULL"
        region = row['region'] if row['region'] != r'\N' else "NULL"
        language = row['language'] if row['language'] != r'\N' else "NULL"
        types = row['types'] if row['types'] != r'\N' else "NULL"
        attributes = row['attributes'] if row['attributes'] != r'\N' else "NULL"
        isOriginalTitle = row['isOriginalTitle'] if row['isOriginalTitle'] != r'\N' else "NULL"
        title = title.replace("'", '"')
        attributes = attributes.replace("'", '"')
        f.write('INSERT INTO aka (title_id, ordering, title, region, language, types, isOriginalTitle, attributes)\n')
        f.write('VALUES ')
        f.write(f"('{title_id}', {ordering}, '{title}','{region}','{language}','{types}','{isOriginalTitle}', '{attributes}');\n")
```
