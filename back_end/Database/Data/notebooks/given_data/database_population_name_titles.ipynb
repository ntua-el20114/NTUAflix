```python
import pandas as pd
```

```python
i = 0
chunk_size = 500000
for chunk in pd.read_csv('../../truncated_data/truncated_title.principals.tsv', sep='\t',chunksize=chunk_size):
    i += 1
    with open(f'../../../SQL/population/done/name_titles_insertions.sql','w') as f:
        for index, row in chunk.iterrows():
            title_id = row['tconst'] if row['tconst'] != r'\N' else "NULL"
            name_id = row['nconst'] if row['nconst'] != r'\N' else "NULL"
            ordering = row['ordering'] if row['ordering'] != r'\N' else "NULL"
            category = row['category'] if row['category'] != r'\N' else "NULL"
            category = category.replace("'",'"')
            characters = row['characters'] if row['characters'] != r'\N' else "NULL"
            characters = characters.replace("'",'')
            characters = characters.replace("[","")
            characters = characters.replace("]","")
            characters = characters.replace('"',"")
            f.write('INSERT INTO principals (title_id, name_id, ordering, category, characters)\n')
            f.write('VALUES ')
            f.write(f"('{title_id}', '{name_id}', {ordering}, '{category}', '{characters}');\n")
```
