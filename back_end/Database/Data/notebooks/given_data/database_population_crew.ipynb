```python
import pandas as pd
crew = pd.read_csv("../../truncated_data/truncated_title.crew.tsv", sep='\t')
```

```python
with open('../../../SQL/population/done/crew_insertions.sql','w') as f:
    for index, row in crew.iterrows():
        title_id = row[0] if row[0] != r'\N' else "NULL"
        directors = row[1].split(',')
        writers = row[2].split(',')
        for director in directors:
            for writer in writers:
                director = director if director != r'\N' else "NULL"
                writer = writer if writer != r'\N' else "NULL"
                f.write('INSERT INTO crew (title_id, director_id, writer_id)\n')
                f.write('VALUES ')
                f.write(f"('{title_id}', '{director}', '{writer}');\n")
```
