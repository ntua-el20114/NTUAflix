```python
import pandas as pd
```

```python
ratings = pd.read_csv("../../truncated_data/truncated_title.ratings.tsv", sep='\t')
```

```python
with open('../../../SQL/population/done/imdb_ratings_insertions.sql','w') as f:
    for index, row in ratings.iterrows():
        title_id = row['tconst'] if row['tconst'] != r'\N' else "NULL"
        averageRating = row['averageRating'] if row['averageRating'] != r'\N' else "NULL"
        numVotes = row['numVotes'] if row['numVotes'] != r'\N' else "NULL"
        f.write('INSERT INTO imdb_rating (title_id, averageRating, numVotes)\n')
        f.write('VALUES ')
        f.write(f"('{title_id}', {averageRating}, {numVotes});\n")
```
