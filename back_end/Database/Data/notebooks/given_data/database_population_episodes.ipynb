```python
import pandas as pd
episodes = pd.read_csv("../../truncated_data/truncated_title.episode.tsv", sep='\t')
```

```python
episodes
```

```python
with open('../../../SQL/population/done/episode_insertions.sql','w') as f:
    for index, row in episodes.iterrows():
        id = row['tconst'] if row['tconst'] != r'\N' else "NULL"
        parent_id = row['parentTconst'] if row['parentTconst'] != r'\N' else "NULL"
        seasonNumber = row['seasonNumber'] if row['seasonNumber'] != r'\N' else "NULL"
        episodeNumber = row['episodeNumber'] if row['episodeNumber'] != r'\N' else "NULL"
        f.write('INSERT INTO episodes (id, parent_id, seasonNumber, episodeNumber)\n')
        f.write('VALUES ')
        f.write(f"('{id}','{parent_id}', {seasonNumber}, {episodeNumber});\n")
```
