```python
i = 0
chunk_size = 500000
for chunk in pd.read_csv('../../truncated_data/truncated_name.basics.tsv', sep='\t',chunksize=chunk_size):
    i += 1
    if i == 2:
        break
    with open(f'../../../SQL/population/done/name_insertions/name_insertions_{i}.sql','w') as f:
        for index, row in chunk.iterrows():
            id = row['nconst'] if row['nconst'] != r'\N' else "NULL"
            primaryName = row['primaryName'] if row['primaryName'] != r'\N' else "NULL"
            primaryName = primaryName if isinstance(primaryName,str) else "NULL"
            primaryName = primaryName.replace("'", '"')
            birthYear = row['birthYear'] if row['birthYear'] != r'\N' else "NULL"
            deathYear = row['deathYear'] if row['deathYear'] != r'\N' else "NULL"
            url = row['img_url_asset'] if row['img_url_asset'] != r"\N" else "NULL"
            url = url.replace("{width_variable}", "original")
            primaryProfession = row['primaryProfession'] if row['primaryProfession'] != r'\N' else "NULL"
            f.write('INSERT INTO name (name_id, primaryName, birthYear, deathYear, primaryProfession, url)\n')
            f.write('VALUES ')
            f.write(f"('{id}', '{primaryName}', {birthYear}, {deathYear},'{primaryProfession}','{url}');\n")
```
